```python
import csv
import os
import matplotlib.lines as mlines

def parseTrace(filename):
    data = []
    with open(filename, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')

        for row in spamreader:
    #         temp = (row[0].split(','))
    #         temp[1] = float(temp[1].replace("?", "-"))
            data.append(float(row[0]))
    
    return data
```

```python
def totalPower(data):
    var = 0
    for i in data:
        var+=i
    return var
```

```python
data = parseTrace('./Pi2-Measure/Pi2.ptrace')

x = [i for i in range(0,len(data))]
len(data), len(x)
idlePowerAverage = 0
for i in data:
    idlePowerAverage+=i
idlePowerAverage = idlePowerAverage/len(data)
idlePowerAverage
```
