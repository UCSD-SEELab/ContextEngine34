```python
import csv
import os
```

```python
data = []
with open('hveravellir-1-jan-1972-31-dec-19.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    
    for row in spamreader:
        temp = (row[0].split(','))
        temp[1] = float(temp[1].replace("?", "-"))
        data.append(temp)
#         print(temp)
```
